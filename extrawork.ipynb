```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('startup_funding.csv')
df.head()
```

```python
sorted(df['Investors Name'].unique().tolist())
```

```python
df['Investors Name'][df['Investors Name'].isnull()]
```

```python
df['Investors Name'] = df['Investors Name'].fillna('Undisclosed')
```

```python
df['Investors Name'].unique()
```

```python
df.info()
```

```python
df.drop(columns=['Remarks'], inplace=True)
```

```python
df.set_index('Sr No', inplace=True)
```

```python
df.rename(columns={'Date dd/mm/yyyy': 'date', 'Startup Name': 'startup', 'Industry Vertical':'vertical','SubVertical':'subvertical', 'City  Location': 'city', 'Investors Name': 'investors', 'InvestmentnType': 'round', 'Amount in USD': 'amount' }, inplace=True)
```

```python
df
```

```python
df['amount'] = df['amount'].fillna(0)
df.info()
```

```python
df['amount'] = df['amount'].str.replace(',','')
df['amount'] = df['amount'].str.replace('undisclosed',"0")
df['amount'] = df['amount'].str.replace('Undisclosed',"0")
df['amount'] = df['amount'].str.replace('unknown',"0")
df['amount'] = df['amount'].str.replace(' ',"")
```

```python
df = df[df['amount'].astype(str).str.isdigit()]
```

```python
df['amount'] = df['amount'].astype(float)
```

```python
def to_miliion(dollar):
    return dollar/1000000
```

```python
df.info()
# df['date'].dt.month
```

```python
df['amount'] = df['amount'].apply(to_miliion)
```

```python
df.tail(25)
```

```python
df['date'] = df['date'].str.replace('05/072018', '05/07/2018')
df['date'] = df['date'].str.replace('01/07/015', '01/07/2015')
df['date'] = df['date'].str.replace('12/05.2015', '12/05/2015')
df['date'] = df['date'].str.replace('13/04.2015', '13/04/2015')
df['date'] = df['date'].str.replace('15/01.2015', '15/01/2015')
df['date'] = df['date'].str.replace('22/01//2015', '22/01/2015')
```

```python
df['date'] = pd.to_datetime(df['date'],format='%d/%m/%Y', dayfirst=True)
```

```python
df.info()
```

```python
df = df.dropna(subset=['date', 'startup', 'vertical', 'city', 'investors', 'round', 'amount'])
```
